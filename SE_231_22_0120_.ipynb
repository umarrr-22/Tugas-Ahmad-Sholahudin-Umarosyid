```python
import webbrowser

website = input("Search Website (google/Elearning Universitas Semarang): ")

if website.lower() == "google":
    query = input("Search: ")
    search_url = f"https://www.google.com/search?q={query}"
    webbrowser.open(search_url)

elif website.lower() == "Elearning":
    query = input("Search: ")
    search_url = f"https://elearning.usm.ac.id//results?search_query={query}"
    webbrowser.open(search_url)

else:
    print("Ketikkan google/Elearning Universitas Semarang aja")
```
